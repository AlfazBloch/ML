```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

img = cv2.imread("0to9.webp",0)
img = cv2.bitwise_not(img)


kernel = np.ones((3,3),np.uint8)

eroision = cv2.erode(src=img,kernel=kernel,iterations=2)

dilation = cv2.dilate(img,kernel,iterations=1)

plt.figure(figsize=(10,3))
plt.subplot(1,3,2),plt.imshow(eroision,cmap='gray'),plt.title('Eroision Effect'),plt.axis('off')
plt.subplot(1,3,3),plt.imshow(dilation,cmap='gray'),plt.title('Dilation Effect'),plt.axis('off')
plt.subplot(1,3,1),plt.imshow(img,cmap='gray'),plt.title('Original'),plt.axis('off')
```

```python
img = cv2.imread('rose2.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)
y,Cr,Cb = cv2.split(img)

plt.figure(figsize=(10,8))
plt.subplot(2,2,1),plt.imshow(cv2.cvtColor(img,cv2.COLOR_YCrCb2RGB)),plt.title("Original"),plt.axis('off')
plt.subplot(2,2,2),plt.imshow(y,cmap='gray'),plt.title("Y Luminanace"),plt.axis('off')
plt.subplot(2,2,3),plt.imshow(Cr,cmap='gray'),plt.title("Chrominance red"),plt.axis('off')
plt.subplot(2,2,4),plt.imshow(Cb,cmap='gray'),plt.title("Chrominance blue"),plt.axis('off')
```
