```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("demo.webp")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.axis('off')
```

```python
gauss = cv2.GaussianBlur(img,(5,5),0)
sharped = cv2.addWeighted(img,1.65,gauss,-.4,0)
plt.imshow(sharped),plt.axis('off')
```

```python
k1 = np.array([[-1,-1,-1],[-1,9.5,-1],[-1,-1,-1]])
k1output = cv2.filter2D(img,-1,k1)
plt.imshow(k1output,cmap="gray"),plt.axis("off")
```

```python
k2 = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
k2output = cv2.filter2D(img,-1,k2)
plt.imshow(k2output,cmap="gray"),plt.axis('off')
```

```python

img = np.zeros((400,600,3))
img = cv2.rectangle(img,(50,50),(550,350),(255),thickness=2)
blur = cv2.blur(img,(5,5))
plt.imshow(blur)
```
