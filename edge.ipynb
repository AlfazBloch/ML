```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('zebra.webp')
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,10)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,10)
sobel = cv2.magnitude(sobelx,sobely)
plt.subplot(2,2,1)
plt.imshow(img,cmap="gray")
plt.subplot(2,2,2)
plt.imshow(sobelx,cmap='gray')
plt.subplot(2,2,3)
plt.imshow(sobely,cmap='gray')
plt.subplot(2,2,4)
plt.imshow(sobel,cmap='gray')
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('zebra.webp')
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gaussian = cv2.GaussianBlur(img,(5,5),0)
laplacian = cv2.Laplacian(img,cv2.CV_64F,ksize=3)
laplacian_of_blur_img = cv2.Laplacian(img,cv2.CV_64F,ksize=3)
plt.figure(figsize=(12,12))
plt.subplot(2,2,1),plt.imshow(img,cmap='gray'),plt.axis('off'),plt.title('zebra image')
plt.subplot(2,2,2),plt.imshow(laplacian,cmap='gray'),plt.axis('off'),plt.title('laplacian image')
plt.subplot(2,2,3),plt.imshow(laplacian_of_blur_img,cmap='gray'),plt.axis('off'),plt.title('laplacian_of_blur_img image')
plt.subplot(2,2,4),plt.imshow(gaussian,cmap='gray'),plt.axis('off'),plt.title('gaussian image')
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('pyramid.webp')
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

gaussian = cv2.GaussianBlur(img,(5,5),1.4)

canny = cv2.Canny(img,threshold1=50,threshold2=150)
canny_of_blur_img = cv2.Canny(gaussian,threshold1=50,threshold2=150)
plt.subplot(2,2,1),plt.imshow(img,cmap='gray'),plt.title("image"),plt.axis('off')
plt.subplot(2,2,2),plt.imshow(gaussian,cmap='gray'),plt.title("gaussian image"),plt.axis('off')
plt.subplot(2,2,3),plt.imshow(canny,cmap='gray'),plt.title("Canny image"),plt.axis('off')
plt.subplot(2,2,4),plt.imshow(canny_of_blur_img,cmap='gray'),plt.title("Canny_of_blur_img image"),plt.axis('off')
```
