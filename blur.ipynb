```python
#box blur

import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('tree.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
blur = cv2.blur(img,(5,5))
plt.subplot(1,2,1)
plt.imshow(img)
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(blur)
plt.axis('off')
```

```python
#Gaussian Blur

import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('tree.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
gaussian = cv2.GaussianBlur(img,(11,11),0)

plt.imshow(gaussian)
plt.title("Gaussian Blur Image")
plt.axis('off')
```

```python
#Median Blur

import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('tree.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
median = cv2.medianBlur(img,5)

plt.imshow(median)
plt.title("Median blur image")
plt.axis('off')
```

```python
#Motion Blur

import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('tree.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
size = 10

# i) Horizontal

k1 = np.zeros((size,size))
k1[size//2,:] = np.ones(size)
k1 = k1/size

horizontal = cv2.filter2D(img,-1,k1)

plt.subplot(1,2,1),plt.title("Horizontal Blur"),plt.imshow(horizontal)
plt.axis('off')


#ii) vertical

k2 = np.zeros((size,size))
k2[:,size//2] = np.ones(size)
k2 = k2 / size

vertical = cv2.filter2D(img,-1,k2)

plt.subplot(1,2,2),plt.title("Vertical Blur"),plt.imshow(vertical)

plt.axis('off')
```
