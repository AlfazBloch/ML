```python
#Contrast effect in gray scale image

import cv2
import matplotlib.pyplot as plt

img = cv2.imread('flower.jpg',cv2.IMREAD_GRAYSCALE)
img1 = cv2.equalizeHist(img)

plt.figure(figsize=(7,5))

plt.subplot(1,2,1),plt.imshow(img,cmap='gray'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(img1,cmap='gray'),plt.axis('off')
```

```python
img = cv2.imread('rose.webp')
ycrcb = cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)

y,cr,cb = cv2.split(ycrcb)

y_eq = cv2.equalizeHist(y)
ycrcb_eq = cv2.merge([y_eq,cr,cb])

global_eq = cv2.cvtColor(ycrcb_eq,cv2.COLOR_YCrCb2BGR)
plt.figure(figsize=(7,5))

plt.subplot(1,2,1),plt.imshow(img),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(global_eq),plt.axis('off')
```

```python
clahe = cv2.createCLAHE(clipLimit=2.0,tileGridSize=(8,8))
y_clahe = clahe.apply(y)

ycrcb_clahe = cv2.merge([y_clahe,cr,cb])

clahe_eq = cv2.cvtColor(ycrcb_clahe,cv2.COLOR_YCrCb2BGR)
plt.figure(figsize=(7,5))

plt.subplot(1,2,1),plt.imshow(img),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(clahe_eq),plt.axis('off')
```
